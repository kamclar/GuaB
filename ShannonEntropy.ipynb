## Functions

```python
def shannon_entropy(seq):
    """
        H(S) = −Σ P(Si) log2 (P(Si))
        P(Si) is the relative frequency of given character in the string.
    """
    
    entropy = 0
    aminoacids = {'G','P','A','V','L','I','M','C','F','Y','W','H','K','R','Q','N','E','D','S','T'}
    for aa in aminoacids:
        rel_freq = seq.count(aa) / len(seq)
        if rel_freq > 0:
            entropy = entropy - (rel_freq * math.log(rel_freq, 2))
        
    return entropy


def get_shannon_sequence(sequence, fn, ws):
    se_sequence = []
    for window in get_window(sequence, ws):
        se = fn(window)
        se_sequence.append(se)
        
    return se_sequence

def get_window(sequence, window_size=10, step=1):
    start = 0
    end = window_size
    
    while end < len(sequence):
        yield sequence[start:end]
        
        start = start + step
        end = end + step

def plot_seqs(shan_seqs_list, names):
    colors = ['b', 'g', 'orange']
    fig = plt.figure(figsize=(20,12))
    for i in range(len(shan_seqs_list)):
        plt.subplot(len(shan_seqs_list),1,i+1)
        plt.plot(shan_seqs_list[i], color=colors[i])
        plt.ylabel("Shannon Entropy")
        plt.title("Shannon Entropy of " + names[i], fontsize=18)        
    plt.xlabel("position", fontsize=12)
    plt.show()
    
def plot_all_seqs(shan_seqs_list, name):
    fig = plt.figure(figsize=(20,5))
    plt.plot(shan_seqs_list[0], color='blue')
    plt.plot(shan_seqs_list[1], color='green')
    plt.plot(shan_seqs_list[2], color='orange')
    plt.xlabel("position", fontsize=12)
    plt.ylabel("Shannon Entropy", fontsize=12)
    plt.title("Shannon Entropy of " + name, fontsize=18)
    plt.show()
```
